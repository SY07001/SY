```
library(ggplot2)
```

```
NBA1516 <- fetch_NBAPlayerStatistics("15-16")
```

```
qplot(FieldGoalsMade, TotalPoints, data = NBA1516) ##使用ggplot畫出x軸是投籃數與y軸是總分之散佈圖
```

```
qplot(FieldGoalsMade, TotalPoints, data = NBA1516,
   geom=c("point","smooth"))
```

```
qplot(FieldGoalsMade, data = NBA1516,fill=Position) ##使用ggplot畫出X軸是投籃數與Y軸是總分之長條圖
```

```
df<-data.frame(gp = factor(rep(letters[1:3],each=10)),y=rnorm(30))
ggplot(df,aes(x=gp, y=y))+geom_boxplot() ##使用ggplot畫出X軸是GP，Y軸是總分盒狀圖
```

```
df$z<-df$y+rnorm(30)
```

```
ggplot(df,aes(x=z,y=y,color=gp))+geom_line() ##使用ggplot畫出Z軸是GP，Y軸是總分折線圖
```
